```python
index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "cloud_provider": {"type": "text"},
            "category": {"type": "text"},
            "service": {"type": "text"},
            "description": {"type": "keyword"} 
        }
    }
}
index_name = "asset-categories"
es_client.indices.delete(index=index_name, ignore_unavailable=True)
es_client.indices.create(index=index_name, body=index_settings)
```

```python
documents[0]
```

```python
from tqdm.auto import tqdm
```

```python
for doc in tqdm(documents):
    es_client.index(index=index_name, document=doc)
```

```python
def elastic_search(query):
    search_query = {
        "size": 10,
        "query": {
            "bool": {
                "must": {
                    "multi_match": {
                        "query": query,
                        "fields": ["description^8", "category^5", "service", "cloud_provider"],
                        "type": "best_fields"
                    }
                },
            }
        }
    }

    response = es_client.search(index=index_name, body=search_query)
    
    result_docs = []
    
    for hit in response['hits']['hits']:
        result_docs.append(hit['_source'])
    
    return result_docs
```

```python
query = 'An interactive query service that makes it easy to analyze data in Amazon S3 using standard SQL.'
results = elastic_search(query)
results
```

```python
def build_prompt(query, search_results):
    prompt_template = """
You're a cloud asset category finder. Answer the QUESTION based on the CONTEXT.
Use only the facts from the CONTEXT when answering the QUESTION. Generate a short answer only from the `answer` field by saying `YOUR CATEGORY IS: ` .

QUESTION: {question}

CONTEXT: 
{context}
""".strip()

    context = ""
    
    for doc in search_results:
        context = context + f"answer: {doc['category']}\nservice: {doc['service']}\ndescription: {doc['description']}\ncloud_provider: {doc['cloud_provider']}\n\n"
    
    prompt = prompt_template.format(question=query, context=context).strip()
    return prompt
```

```python
print(build_prompt(query, results))
```

```python
def flant5_llm(prompt):
    input_text = prompt
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids

    outputs = model.generate(input_ids)
    print(tokenizer.decode(outputs[0]))
```

```python
def rag_flant5(query):
    search_results = elastic_search(query)
    prompt = build_prompt(query, search_results)
    answer = flant5_llm(prompt)
    return answer
```

```python
from openai import OpenAI
client = OpenAI(base_url='http://localhost:11434/v1/', api_key='ollama')
client._version
```

```python
def ollama_llm(prompt):
    response = client.chat.completions.create(
        model='gemma2:2b',
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content
```

```python
def rag(query):
    search_results = elastic_search(query)
    prompt = build_prompt(query, search_results)
    answer = ollama_llm(prompt)
    return answer
```

```python
query = 'A query generator which allows you to run multiple queries to fetch the data'
rag(query).strip()
```
